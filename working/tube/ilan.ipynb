```python
import numpy as npy
from matplotlib import pyplot

ga=1.4
t=0.01
dt=0.002 #time step
dx=0.25 #space step
nx=81 #number space division
nt=int((t/dt))+1 #number time disivion
x=npy.linspace(-10,10,nx)

u=npy.zeros((3,nx))

up=npy.zeros((3,nx))
up[0,int(nx/2):]=0.125
up[2,int(nx/2):]=25000

up[0,:int(nx/2)]=1
up[2,:int(nx/2)]=250000

def f(u,ga):
    f=npy.zeros((3,nx))
    f[0,:]=u[1,:]
    f[1,:]=(u[1,:]**2)+((ga-1)*(u[2,:]-(u[1,:]**2/u[0])))
    f[2,:]= (u[1,:]/u[0,:])*(u[2,:]+(ga-1)*(u[2,:]-((u[1,:]**2)/u[0,:])))
    return f

u12=npy.zeros((3,nx))
un=npy.zeros((3,nx))
u=npy.zeros((3,nx))
u=up.copy()
for n in range(1,nt):
    for j in range(0,2):
        u12[j,:-1]=1/2*(u[j,1:]+u[j,:-1])-(((dt/2*dx)*(f(u,ga)[j,1:]))-f(u,ga)[j,:-1])
        un[j,1:]=u[j,1:]-((dt/dx)*(f(u12,ga)[j,1:]-f(u12,ga)[j,:-1]))
print(un)
```
